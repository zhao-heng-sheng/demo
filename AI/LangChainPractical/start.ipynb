```typescript
import {load} from 'dotenv'
const env = await load()
const process = {env}
```

```typescript
import { ChatOpenAI } from "@langchain/openai";
import { ChatBaiduQianfan } from "@langchain/baidu-qianfan";
import { HumanMessage } from "@langchain/core/messages";
const chatModel = new ChatBaiduQianfan({
  model:"ERNIE-Speed-128K",
  name:"test"
})
await chatModel.invoke([
  new HumanMessage('请讲个笑话')
])
```

```typescript
import { ChatOpenAI } from "@langchain/openai";
import { HumanMessage } from "@langchain/core/messages";
import { StringOutputParser } from "@langchain/core/output_parsers";
const chatModel = new ChatOpenAI({
  configuration:{
    baseURL: process.env.baseURL,
  },
})
const outputPrase = new StringOutputParser()
const simpleChain = chatModel.pipe(outputPrase)
await simpleChain.invoke([
  new HumanMessage('请讲个笑话')
])
```

```typescript
await simpleChain.batch([
  [
    new HumanMessage('34534264*1234等于多少？')
  ],
  [
    new HumanMessage('蛇真的能吞掉大象吗？')
  ],
])
```

```typescript
const stream = await simpleChain.stream(
  [
    new HumanMessage('为什么一直学习会感到烦躁')
  ]
)
for await (const chunk of stream){
  console.log(chunk)
}
```

```typescript
import { ChatOpenAI } from "@langchain/openai";
const fakeLLM = new ChatOpenAI({
  configuration:{
    baseURL: process.env.baseURL,
  },
  openAIApiKey: "123",
  maxRetries:0,
})
const realLLM = new ChatOpenAI({
  configuration:{
    baseURL: process.env.baseURL,
  },
});
const llmWithFallback = fakeLLM.withFallbacks({
  fallbacks: [realLLM]
})
await llmWithFallback.invoke("你好")
```
